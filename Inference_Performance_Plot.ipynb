```python
# data from https://allisonhorst.github.io/palmerpenguins/

import matplotlib.pyplot as plt
import numpy as np

species = ("Adelie", "Chinstrap", "Gentoo")
penguin_means = {
    'Bill Depth': (18.35, 18.43, 14.98),
    'Bill Length': (38.79, 48.83, 47.50),
    'Flipper Length': (189.95, 195.82, 217.19),
}

x = np.arange(len(species))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in penguin_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Length (mm)')
ax.set_title('Penguin attributes by species')
ax.set_xticks(x + width, species)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 250)

plt.show()
```

## Created data visualization chart modified
We are going to use previous template to create our own benchmark performance bar chart:
- **X axis**: The model name in string (e.g. Classification, Object Detection, Segmentation)
- **Y axis**: The benchmark performance test results in tuple (e.g. (100, 200, 300))

Changed **x values** from `species` to `model_name`<br>
Changed **y values** from `penguin_means` to `test_results`

```python
import matplotlib.pyplot as plt
import numpy as np

model_name = ("Classification\nVGG16", "Object Detection\nYOLOv3", "Segmentation\nIntel-iss-0228")
test_results = {
    'CPU': (24.49, 13.87, 5.79),
    'GPU (Arc)': (1790.13, 657.44, 58.62),
}

x = np.arange(len(model_name))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in test_results.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Throughput (FPS)', fontsize=12)
ax.set_title('Benchmark Performance by Models', fontsize=14)
ax.set_xticks(x + width, model_name, fontsize=12)
ax.legend(loc='upper right', ncols=3, fontsize=12)
ax.set_ylim(0, 1000)

plt.savefig('my-ov-cpu-gpu-plot.png',  bbox_inches='tight')

plt.show()
```

```python
model_names = ("Classification\nVGG19", "Object Detection\nYOLOv4", "Segmentation\nU-Net")
test_results = {
    'CPU': (10, 20, 200),
    'GPU (Arc)': (50, 100, 150),
}

plot_performance(model_names, test_results, save_figure=True, saved_name='my-plot-2023-0323.png')
```
